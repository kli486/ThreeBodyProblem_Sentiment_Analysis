```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\WM\340 Final Project\DataSet\Negative_comment.csv")
all_text = ' '.join(df['Negative_comment'])

# Handle Case Sensitive
all_text = all_text.lower()
all_text
```

```python
# Generate the word cloud
custom_stopwords=['book','i','it','the','this','that','and','a','from','is','in','to']
wordcloud = WordCloud(width=800, height=800, background_color='black', min_font_size=10,stopwords=custom_stopwords).generate(all_text)

# Display the word cloud using matplotlib
plt.figure(figsize=(6, 6), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from collections import Counter


df=pd.read_csv(r"C:\WM\340 Final Project\DataSet\Positive_comment.csv")

# Combine all comments into one string
all_comments = ' '.join(df['Positve_comment'])  
tokens = word_tokenize(all_comments)
tags = pos_tag(tokens)

adjectives = [word for word, tag in tags if tag.startswith('JJ')]

top_adjectives = Counter(adjectives).most_common(20)

print(top_adjectives)
```
